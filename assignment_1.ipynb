#  3.Handle missing values (identify, fill/drop).

In this step, missing values are identified and handled using appropriate methods.

```python
df.isnull().sum()
```

```python
df = df.fillna(0)
```

# 4.Remove duplicates

In this step, duplicate rows are identified and removed from the dataset.

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

#  5.Perform basic operations (filter rows, select columns). 

In this step, basic operations such as selecting columns and filtering rows are performed on the dataset.

```python
df[['title', 'final_price']]
```

```python
high_rated = df[df['rating'] > 4]
high_rated.head()
```

# 6.Create a derived column (total_amount = price * quantity).

In this step, a new column named total_amount is created using price and quantity columns.

```python
df['quantity'] = 1
df['final_price'] = df['final_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['total_amount'] = df['final_price'] * df['quantity']
print(df[['final_price', 'quantity', 'total_amount']].head())
```

```python
df.head()
```

#  7: Save the Cleaned Dataset as a New CSV File

In this step, the cleaned dataset is saved as a new CSV file for future analysis.

```python
df.to_csv("cleaned_combined_dataset.csv", index=False)
```
